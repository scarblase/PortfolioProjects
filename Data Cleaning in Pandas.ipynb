```python
import pandas as pd
import seaborn as sls
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Portfolio Projects (Data Analysis)\Python(Projects)\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
if 'Not_Useful_Column' in df.columns:
    df = df.drop(columns='Not_Useful_Column')
else:
    print("Column 'Not_Useful_Column' does not exist in the DataFrame")
```

```python
#df['Last_Name'] = df['Last_Name'].str.lstrip("...")
#df['Last_Name'] = df['Last_Name'].str.lstrip("/")
#df['Last_Name'] = df['Last_Name'].str.rstrip("_")
#df['Last_Name'] = df['Last_Name'].str.lstrip("_")
df['Last_Name'] = df['Last_Name'].str.strip("123./")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r"[^a-zA-Z0-9]", "", regex=True)
df["Phone_Number"] = df["Phone_Number"].astype(str)

df["Phone_Number"] = df["Phone_Number"].apply(
    lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:] if len(x) >= 10 else x
)
df["Phone_Number"] = df["Phone_Number"].str.replace("nan","")
df["Phone_Number"] = df["Phone_Number"].str.replace("Na","")

df["Phone_Number"]
```

```python
df[['Street Address', 'State', 'Zip_Code']] = df['Address'].str.split(",", n=2, expand=True)
df
```

```python
df["Do_Not_Contact"] = ""  # Start with empty strings

df.loc[df["Paying Customer"].str.lower() == "no", "Do_Not_Contact"] = "Yes"
df.loc[df["Phone_Number"].isnull(), "Do_Not_Contact"] = "Yes"

df["Do_Not_Contact"] = df["Do_Not_Contact"].replace("", "No")

print(df[["CustomerID", "First_Name", "Last_Name", "Do_Not_Contact"]])
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == "Yes":
        df.drop(x, inplace=True)

for x in df.index:
    if df.loc[x, "Phone_Number"] == "":
        df.drop(x, inplace=True)
        
df
        
```
