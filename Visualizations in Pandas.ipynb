```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
plt.style.use('Solarize_Light2')
```

```python
df = pd.read_csv("Ice Cream Ratings.csv")
df
```

```python
df.plot(kind='line', title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Scores')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

df.plot(kind='bar', stacked=True, ax=axes[0], title='Vertical Stacked')

df.plot.barh(stacked=True, ax=axes[1], title='Horizontal Stacked')

plt.tight_layout()
plt.show()
```

```python
df.plot.scatter(x='Texture Rating',y = 'Overall Rating', s=50, color='purple')
```

```python
df.plot.hist(bins=30)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10,5), legend=0)
```

```python
df.plot.pie(y='Flavor Rating', figsize=(10,6))
plt.legend(labels=df['Date'], loc='best')
```
